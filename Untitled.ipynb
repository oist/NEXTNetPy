```python
import nmepinet as nmen
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

Simulate $20000$ times an epidemic on a single Barabasi-Albert network of size $N=10^4$. 
We retrieve the quantities $Z_k(n)$ and $P_n(k)$

```python
SIZE=10**4
SIM=20000
SEED=1
pk,knn = nmen.depleted_distribution(SIZE,SIM,SEED)
zk=nmen.simulation_discrete_zk(SIZE,SIM,SEED)
```
