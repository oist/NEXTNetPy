```python
g=nx.configuration_model(degree)
g = nx.Graph(g)
g.remove_edges_from(nx.selfloop_edges(g))
```

```python
psi=nmen.time_distribution(5,3)


times,infected = nmen.simulate(g,psi,initial_infected=100)
```

```python
L=nmen.growth_rate(g,psi)
```
